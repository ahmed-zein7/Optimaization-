# batch

```python
def Batch( x , y , alpha , max_iterations):
    theta=np.zeros((x.shape[1],1))
    theta0_all = []
    theta1_all = []
    loss = []
    m = len(x)

    for i in range(max_iterations):
        h = x @ theta
#         allhypothes.append(h)
        cost = (1/(2*m)) *np.sum(np.square(h-y))
        loss.append(cost)       
        theta0_all.append(theta[0][0])
        theta1_all.append(theta[1][0])
        err=h-y

        gradient_theta =(x.T@err)/m
        theta = theta- alpha * gradient_theta
    

    return h,theta,loss,theta0_all,theta1_all
```

# Mini Batch

```python
def Mini_Batch(x,start,batch_size):
    return x[start:start+batch_size,:]
```

```python
def GD_Mini_Batch( x , y , alpha , itterations,batch_size ):
    theta=np.zeros((x.shape[1],1))
    theta0_all = []
    theta1_all = []
    loss = []
    allhypothes = []
    m = len(x)

    for i in range(itterations):
#X is a matrix with len(X)/batches rows
        for j in range(0 ,  m , batch_size):
            x_mini = Mini_Batch(x,j,batch_size)
            y_mini = Mini_Batch(y,j,batch_size)
            h =  x_mini @ theta
#           allhypothes.append(h)
            cost = (1/(2*batch_size)) * np.sum(np.square(h-y_mini))
            loss.append(cost)       
            theta0_all.append(theta[0][0])
            theta1_all.append(theta[1][0])
            err=h-y_mini

            gradient_theta =(x_mini.T@err)/batch_size
    #       gradient_theta1 = (1/m) *np.sum((h-y)*x)
            theta = theta- alpha * gradient_theta
        
        


    h =  x @ theta    
      
    return h,theta,loss,theta0_all,theta1_all
```

# ADAM_MINI_BATCH

```python
def ADAM_Mini_Batch( x , y , alpha , itterations,batch_size,beta1,beta2 ):
    theta=np.zeros((x.shape[1],1))
    mt=0
    vt=0
    epslon=1e-8
    theta0_all = []
    theta1_all = []
    loss = []
    
    m = len(x)

    for i in range(itterations):
#X is a matrix with len(X)/batches rows
        for j in range(0 ,  m , batch_size):
            x_mini = Mini_Batch(x,j,batch_size)
            y_mini = Mini_Batch(y,j,batch_size)
            h =  x_mini @ theta
            err=h-y_mini
            cost = (1/(2*batch_size)) *np.sum(np.square(h-y_mini))
           
                
            
            loss.append(cost)  
            
            theta0_all.append(theta[0][0])
            theta1_all.append(theta[1][0])
            

            gradient_theta =(x_mini.T@err)/batch_size
            mt = beta1 *mt + (1-beta1) * gradient_theta
            vt = beta2 *vt + (gradient_theta**2)
            
            theta =theta - (alpha  / ( np.sqrt(vt) +epslon) * mt)
           


        if abs(loss[-1]-loss[-2]) < 0.0001 :
            print(f"  it tooks  {i+1} iterations ")
            break
               
        
           

    h =  x @ theta  
      
    return h,theta,loss,theta0_all,theta1_all
```

```python
def RMS_Mini_Batch( x , y , alpha , itterations,beta,batch_size ):
    theta=np.zeros((x.shape[1],1))
    v=0
    theta0_all = []
    theta1_all = []
    loss = []
    allhypothes = []
    m = len(x)
    epslon=1e-8

    for i in range(itterations):
#X is a matrix with len(X)/batches rows
        for j in range(0 ,  m , batch_size):
            x_mini = Mini_Batch(x,j,batch_size)
            y_mini = Mini_Batch(y,j,batch_size)
            h =  x_mini @ theta
#           allhypothes.append(h)
            cost = (1/(2*batch_size)) * np.sum(np.square(h-y_mini))
            loss.append(cost)       
            theta0_all.append(theta[0][0])
            theta1_all.append(theta[1][0])
            err=h-y_mini

            gradient_theta =(x_mini.T@err)/batch_size
            
            
           
            v =  beta * v + (1-beta) *(gradient_theta**2)
            theta =theta - (alpha  /  (np.sqrt(v) +epslon) * gradient_theta)
            



    h =  x @ theta    
      
    return h,theta,loss,theta0_all,theta1_all
```

## NAG 

```python
def NAG_Mini_Batch( x , y , alpha , itterations,batch_size ):
    theta=np.zeros((x.shape[1],1))
    v=np.zeros((x.shape[1],1))
    theta0_all = []
    theta1_all = []
    loss = []
    allhypothes = []
    m = len(x)
    gamma = 0.9

    for i in range(itterations):
#X is a matrix with len(X)/batches rows
        for j in range(0 ,  m , batch_size):
            x_mini = Mini_Batch(x,j,batch_size)
            y_mini = Mini_Batch(y,j,batch_size)
            h =  x_mini @ theta
#           allhypothes.append(h)
            cost = (1/(2*batch_size)) * np.sum(np.square(h-y_mini))
            loss.append(cost)       
            theta0_all.append(theta[0][0])
            theta1_all.append(theta[1][0])
            err=h-y_mini

#             gradient_theta =(x_mini.T@err)/batch_size
    #       gradient_theta1 = (1/m) *np.sum((h-y)*x)
           



            theta_temp = theta - gamma * v
            h_temp = x_mini @ theta_temp 
            

            gradient_theta_temp= (1/batch_size) * np.sum( h_temp - y_mini) 
            theta = theta_temp - alpha * gradient_theta_temp
            v = gamma * v + alpha * gradient_theta_temp
          
        


    h =  x @ theta    
      
    return h,theta,loss,theta0_all,theta1_all
```
